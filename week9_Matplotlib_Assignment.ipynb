# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib. 

Matplotlib is a Python library used for creating static, animated, and interactive visualizations in Python ยน. It provides a MATLAB-like interface and can be used in Python scripts, the Python and IPython shell, web application servers, and various graphical user interface toolkits like Tkinter, awxPython, etc. . 

The Pyplot module of Matplotlib is a collection of functions that provide a simple way to create a variety of charts and plots. Here are five types of plots that can be plotted using the Pyplot module of Matplotlib:

1. **Line Plot**: A line plot is a type of chart that displays information as a series of data points connected by straight line segments .
2. **Histogram**: A histogram is a type of chart that displays the distribution of a dataset. It groups the data into bins and displays the number of data points that fall into each bin .
3. **Scatter Plot**: A scatter plot is a type of chart that displays the relationship between two variables. It shows how much one variable is affected by another .
4. **3D Plot**: A 3D plot is a type of chart that displays data in three dimensions. It is useful for visualizing complex data and finding patterns in the data .
5. **Image Plot**: An image plot is a type of chart that displays an image. It is useful for visualizing images and finding patterns in the data .



# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot. Note: Also add title, xlabel, and ylabel to the plot.



```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

**Scatter Plot**: A scatter plot is a type of chart that displays the relationship between two variables. It shows how much one variable is affected by another .

```python
import matplotlib.pylab as plt
plt.scatter(x,y, c='r')
plt.title('scatter plot')
plt.xlabel("data in x axis")
plt.ylabel("data in y axis")
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

The subplot() function in Matplotlib is used to create multiple plots in one figure. It takes three arguments that describe the layout of the figure: the number of rows, the number of columns, and the index of the current plot 

```python
import numpy as np
x=np.array([0, 1, 2, 3, 4, 5])
y1=np.array([0, 100, 200, 300, 400, 500])
y2=np.array([50, 20, 40, 20, 60, 70])
y3=np.array([10, 20, 30, 40, 50, 60])
y4=np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Plot 1')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Plot 2')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Plot 3')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Plot 4')

plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


A box plot is a type of chart that displays the distribution of a dataset. It shows the median, quartiles, and outliers of the data . Box plots are useful for identifying the range, skewness, and outliers of a dataset .

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
fig, ax = plt.subplots()
box=[box1,box2]
ax.boxplot(box)

# Labels
ax.set_title('Box Plot of Data')
ax.set_xlabel('Data')
ax.set_ylabel('Values')

ax.boxplot(box2)

# Labels
ax.set_title('Box Plot of Data')
ax.set_xlabel('Data')
ax.set_ylabel('Values')


```
