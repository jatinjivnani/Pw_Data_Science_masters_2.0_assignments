# Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_dist(x, mean, sd):
    
    prob_density = np.exp(-0.5 * ((x - mean) / sd) ** 2)/(sd*(2*np.pi)**0.5)
    return prob_density

# Example usage:
x_values = np.linspace(1, 50, 200)  # Creating a series of data points
mean_value = np.mean(x_values)
std_deviation = np.std(x_values)

pdf_values = normal_dist(x_values, mean_value, std_deviation)

# Plotting the PDF
plt.plot(x_values, pdf_values, color='red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
plt.title('Normal Distribution PDF')
plt.show()
```

# Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

Properties of Binomial Distribution:

Two Possible Outcomes:
The Binomial distribution deals with experiments having only two distinct outcomes: success or failure, yes or no, true or false.
These outcomes are often represented as (1) (success) and (0) (failure).

Fixed Number of Trials:
In a Binomial experiment, there is a fixed number of independent trials (denoted as (n)).
Each trial is identical and independent of the others.

Constant Probability of Success:
The probability of success (denoted as (p)) remains constant for each trial.
The probability of failure is (1 - p).

Binomial Random Variable:
The number of successes in (n) repeated trials of a Binomial experiment is represented by a binomial random variable.
This random variable can take values from (0) to (n).

Examples of Binomial Distribution Applications:
Number of Side Effects from Medications
tossing a coin n times

# Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random sample from a binomial distribution
n_trials = 1000
p_success = 0.4
sample = np.random.binomial(n=1, p=p_success, size=n_trials)  # 1 indicates success (yes), 0 indicates failure (no)

# Plot a histogram
plt.hist(sample)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=1000, p=0.4)')
plt.show()
```

# Q7. How Binomial distribution different from Poisson distribution?


1. **Definition**:
   - **Binomial Distribution**:
     - The Binomial distribution deals with experiments having only **two distinct outcomes**: success or failure, yes or no, true or false.
     - It is based on a **fixed number of independent trials** (denoted as \(n\)) where each trial has a constant probability of success (\(p\)).
   - **Poisson Distribution**:
     - The Poisson distribution describes the probability of a **certain number of events happening in a fixed time interval**.
     - It is based on **continuous events** rather than fixed trials.

2. **Nature**:
   - **Binomial Distribution**:
     - Biparametric (characterized by two parameters: \(n\) and \(p\)).
     - Only two possible outcomes: success or failure.
     - Success probability remains constant.
   - **Poisson Distribution**:
     - Uniparametric (characterized by one parameter: \(\mu\)).
     - An infinite number of possible outcomes (from \(x = 0\) to \(x = \infty\)).

3. **Examples**:
   - **Binomial Distribution**:
     - Coin tossing experiments (heads or tails).
     - Printing mistakes per page of a large book.
   - **Poisson Distribution**:
     - Counting occurrences of rare events (e.g., earthquakes, phone calls, accidents).
     - Modeling the number of customers arriving at a store in a given hour.

4. **Mean and Variance**:
   - **Binomial Distribution**:
     - Mean (\(\mu\)) is greater than variance (\(\sigma^2\)): \(\mu = np\), \(\sigma^2 = npq\).
   - **Poisson Distribution**:
     - Mean (\(\mu\)) equals variance: \(\mu = \sigma^2\).



# Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np
from scipy.stats import poisson

# Generate a random sample from a Poisson distribution
sample_size = 1000
mean_value = 5
sample = np.random.poisson(mean_value, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print(f"Sample Mean: {sample_mean:}")
print(f"Sample Variance: {sample_variance:}")
```

# Q9. How mean and variance are related in Binomial distribution and Poisson distribution?



1. **Poisson Distribution**:
   - In the Poisson distribution, both the mean (average) and variance are **equal** and are denoted by the parameter \(\lambda\) (lambda).
   - This property of equal mean and variance is a distinctive characteristic of the Poisson distribution and simplifies its statistical analysis.
   - Mathematically, for a Poisson distribution:
     \[ E(X) = \lambda \quad \text{(Mean)} \]
     \[ \text{Var}(X) = \lambda \quad \text{(Variance)} \]

2. **Binomial Distribution**:
   - For the Binomial distribution, the relationship between mean and variance is different:
     - The mean (\(\mu\)) is **larger** than the variance (\(\sigma^2\)).
     - Mathematically, for a Binomial distribution:
       \[ E(X) = np \quad \text{(Mean)} \]
       \[ \text{Var}(X) = npq \quad \text{(Variance)} \]
     - Here, \(n\) represents the number of trials, \(p\) is the probability of success, and \(q = 1 - p\) is the probability of failure.

3. **Summary**:
   - In summary:
     - **Poisson distribution**: Mean and variance are **equal** (\(E(X) = \text{Var}(X)\)).
     - **Binomial distribution**: Mean is **greater** than variance (\(E(X) > \text{Var}(X)\)).



# Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

The least frequent data in a normal distribution appears in the tails, far away from the mean. These extreme values have lower probabilities of occurrence compared to the values near the center of the distribution.
