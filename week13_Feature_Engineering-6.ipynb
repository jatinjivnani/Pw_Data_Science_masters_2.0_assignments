# Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

```python
import pandas as pd
import numpy as np
import seaborn as sns
np.random.seed(26)
Hours=[9,8,7,5,3,2,4,6,7,9,10,12,4,5,1,2,3,4,1,3,6,7,8]
marks=[80,75,79,60,20,88,35,60,12,92,100,88,1,0,2,20,15,32,45,76,89,65,99]
df=pd.DataFrame({'hours':Hours,
                'marks':marks})
df.corr(method='pearson')
sns.heatmap(df.corr(method='pearson'),annot=True)
```

```python
sns.pairplot(df.corr(method='pearson'))
```

## as we can see from above by increasing number of hours. student marks also increasing


# Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.

```python
import pandas as pd
import numpy as np
import seaborn as sns
np.random.seed(26)
sleep=[9,8,7,5,3,2,4,6,7,9,10,12,4,5,1,2,3,4,1,3,6,7,8]
satisfaction=[8,7,9,0,2,8,5,4,2,9,10,8,1,0,2,0,5,3,5,6,8,3,8]
df=pd.DataFrame({'sleep':sleep,
                'satisfaction':satisfaction})
df.corr(method='spearman')
sns.heatmap(df.corr(method='spearman'),annot=True)
```

## there is possitive corelation between sleep and satisfaction. we can say that as amount of sleep increases satisfaction increases

# Q6. A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week. The company collected data on both variables from a sample of 30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

```python
import pandas as pd
import numpy as np
import seaborn as sns
# np.random.seed(4)
call_per_day=np.random.randint(low=0,high=100,size=30)
sales_per_week=np.random.randint(low=0,high=100,size=30)
df=pd.DataFrame({'call_per_day':call_per_day,
                'sales_per_week':sales_per_week})
df.corr(method='pearson')
sns.heatmap(df.corr(method='pearson'),annot=True)
```

## possitive correlated. as number of call per day increase the number of sales per week also increase
