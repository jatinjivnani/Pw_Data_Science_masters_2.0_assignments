## Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

`loc` and `iloc` are two methods in Pandas DataFrame that are used for selecting data from the DataFrame. The main difference between the two is the way they are used to select data.

`loc` is a label-based method that selects data based on the labels of rows or columns. It accepts the label of the row or column as input and returns the data associated with that label. `loc` includes the last element of the range passed in it, unlike `iloc`. `loc` can accept boolean data unlike `iloc`. 

`iloc` is an integer-based method that selects data based on the integer position of rows or columns. It accepts the integer position of the row or column as input and returns the data associated with that position. `iloc` does not include the last element of the range passed in it, unlike `loc`.

Here is a table summarizing the differences between `loc` and `iloc`:

| **Method** | **Selection Based On** | **Includes Last Element** |
|------------|-----------------------|---------------------------|
| loc        | Label                 | Yes                       |
| iloc       | Integer Position      | No                        |

## Q6. What do you understand about the windows function in pandas and list the types of windows functions?

`pandas` provides a set of APIs for performing windowing operations, which perform an aggregation over a sliding partition of values. There are four types of windowing operations supported by `pandas`:

1. **Rolling window**: A generic fixed or variable sliding window over the values.
2. **Weighted window**: A weighted, non-rectangular window supplied by the `scipy.signal` library.
3. **Expanding window**: An accumulating window over the values.
4. **Exponentially Weighted window**: An accumulating and exponentially weighted window over the values.

The `rolling` method is used to perform a rolling window operation on a `pandas` object. The `rolling` method returns a `Rolling` object, which can be used to perform various windowing operations. The `rolling` method accepts a `window` parameter, which specifies the size of the window. The `window` parameter can be an integer, which specifies the number of rows to include in the window, or a string, which specifies a time offset.

Here is an example of how to use the `rolling` method to perform a rolling window operation on a `pandas` DataFrame:

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3, 4, 5]})

# Perform a rolling window operation
rolling_window = df['A'].rolling(window=2)

# Calculate the rolling mean
rolling_mean = rolling_window.mean()

# Print the result
print(rolling_mean)
```

This will output:

```
0    NaN
1    1.5
2    2.5
3    3.5
4    4.5
Name: A, dtype: float64
```

## Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

````python
import pandas as pd

# Prompt the user to enter the dates
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

# Convert the dates to datetime objects
date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)

# Calculate the time delta between the dates
delta = date2 - date1

# Extract the number of days, hours, and minutes from the time delta
days = delta.days
hours = delta.seconds // 3600
minutes = (delta.seconds % 3600) // 60

# Display the result
print(f"The difference between {date1.date()} and {date2.date()} is {days} days, {hours} hours, and {minutes} minutes.")
````

## Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.

````python
 import pandas as pd

# Prompt the user to enter the file path
file_path = input("Enter the file path: ")

# Read the CSV file into a DataFrame

df = pd.read_csv(file_path)

# Prompt the user to enter the column name
column_name = input("Enter the column name: ")

# Prompt the user to enter the category order
category_order = input("Enter the category order (comma-separated): ").split(",")

# Convert the specified column to a categorical data type
df[column_name] = pd.Categorical(df[column_name], categories=category_order)

# Sort the DataFrame by the specified column

df = df.sort_values(by=column_name)

# Display the sorted data

df
````

## Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and displays the results in a table.

````python
import pandas as pd
file_path = input("Enter the file path: ")

# Read CSV file
df = pd.read_csv(file_path)
mean=df['score'].mean()
median=df['score'].median()
mode=df['score'].mode()
results = pd.DataFrame({'Statistic': ['Mean', 'Median', 'Mode'],
                        'Value': [mean, median, mode[0]]})
results
````
